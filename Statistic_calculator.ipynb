```python
import numpy as np
```

```python
def stat_cal(l):
  l = np.array(l, dtype=float)
  return{
      "Mean"     : np.mean(l),
      "Median"   : np.median(l),
      "Std div"  : np.std(l),
      "Variance" : np.var(l),
      "minimum"  : np.min(l),
      "maximum"  : np.max(l),
      "Range"    : np.ptp(l),
      "Sum"      : np.sum(l)
  }
```

```python
n=input("Enter the numbers: \n")
l=list(map(float,n.strip().split()))
if len(l)<2:
  print("Enter atleast two numbers")
else:
  result=stat_cal(l)
  for k,v in result.items():
    print(f" {k:<12} : {v:.4f}")
```
